```python
def define_kind_paper(file_ID, THREAD):
    
    global list_kind
    
    for i in file_ID:

        html_file = open('/Users/cristinamenghini/Desktop/EPFL/Fall semester/Applied data analysis - ADA/Data_modERN/infoscience/' + i,'r').read()

        soup = BeautifulSoup(html_file, 'lxml')
        # Get the kind of article
        types = soup.find_all(attrs={"class": "doctype local-color"})
        if len(types) != 0:
            list_kind += [types[0].text]
    print ("Thread n", THREAD, 'DONE')
```

```python
def files_chunks(n, file_list):
    """This function returns the list of chunks that will be used during the multi-threading.
    @n: is the number of threads;
    @file_list: is the list of user to split."""
    num = float(len(file_list))/n 
    f_lists = [ file_list[i:i + int(num)] for i in range(0, (n-1)*int(num), int(num))]
    f_lists.append(file_list[(n-1)*int(num):])
    return f_lists

def worker(file_ID, THREAD):
    """This function define the work that each thread is going to do.
    The variables of the function are the arguments of Get_followers function."""
    define_kind_paper(file_ID, THREAD)
```

```python
file_lists = files_chunks(20, os.listdir('/Users/cristinamenghini/Desktop/EPFL/Fall semester/Applied data analysis - ADA/Data_modERN/infoscience/')[1:] )
```

```python
#Initialize the entire list of edges
list_kind = []
```

```python
s = time.time()
threads = []
for j in range(len(file_lists)):
    t = threading.Thread(target = worker, args = (file_lists[j],j,)) # That's the funcion that each thread will execute
    threads.append(t)
    time.sleep(1)
    print ("Thread", j)
    t.start()
print (time.time - t)
```
